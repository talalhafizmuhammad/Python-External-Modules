# Matplotlib

#### A library useful to create visualized data stats, plotting graphs and more...

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # importing matplotlib
```

### Matplotlib basics

```python
# creating two numpy arrays x and y from .linspace()

x = np.linspace(0, 5, 11)
x
y = x ** 3
y
```

```python
plt.plot(x, y) # Basic line graph formattion from matplotlib
```

```python
# Customizing my line graph
plt.plot(x, y)
plt.title('My Graphical Analysis')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

### Sub-Plots in Matplotlib

```python
# Creating a 2 x 2 subplot, having 4 boxes... and plotting some values in it. 

plt.subplot(2, 2, 1)
plt.plot(x, x)

plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.subplot(2, 2, 4)
plt.plot(x**2, x**3)

plt.subplot(2, 2, 2)
plt.plot(y, x)
```

```python
# Creating a 6 x 6 plot, not useful in all cases, 2 x 2 is enough, since 6 x 6 or others will not be filled properly in this canvas

plt.subplot(6, 6, 1)
plt.subplot(6, 6, 2)
plt.subplot(6, 6, 3)
plt.subplot(6, 6, 4)
plt.subplot(6, 6, 6)
plt.subplot(6, 6, 11)
plt.subplot(6, 6, 21)
plt.subplot(6, 6, 35)
```

### Object-Oriented Way

##### We can not plot my one plot to other oner, so, we will use OOW
- White background as a whole, in subplots is known as canvas
- We can not select the canvas as a whole


```python
fig = plt.figure(figsize = (8, 8), dpi = 100)
axes = fig.add_axes([0.1, 0.7, 0.8, 0.8])
axes.plot(x, y)

axes2 = fig.add_axes([0.3, 0.8, 0.5, 0.5])
axes2.plot(x, y)
```

Matplotlib allows the aspect ratio, figure size and DPI to be specified when figire object is
created.
You can use figsize and DPI keyword arguments.
figsize is the tuple specifying the width, and height of the figure in inches.
DPI is the dots-per-inches, we know.
