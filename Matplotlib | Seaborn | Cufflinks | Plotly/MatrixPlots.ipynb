# Matrix Plots

- Heat map

- Cluster map
- Pivot table heat map

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

### Heat Maps

For heatmap, one must know
- Co-relation

```python
tipsCorr = tips[['total_bill', 'tip', 'size']]
tipsCorr
```

```python
map = tipsCorr.corr()# Will show the co-relation between the terminologies, like upto which extent tip and size can have the relation
map
```

```python
sns.heatmap(map)
```

```python
sns.heatmap(map, annot = True)
```

### Cluster Map

```python
sns.clustermap(map, annot = True)
```

### Pivot table heat map

```python
pvtflight = flights.pivot_table(values = 'passengers', index = 'month', columns = 'year')
```

```python
pvtflight
```

```python
pvt = sns.heatmap(pvtflight)
```
