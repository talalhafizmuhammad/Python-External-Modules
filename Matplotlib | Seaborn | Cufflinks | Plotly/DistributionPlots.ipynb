! Important content: When processing a data, it is categorized as:
- Qualitative data (categorized data, choices for user... like M or F)
- Quantitative data (Distributed form, like age, different for all)


## Distribution Plots

##### We use distribution plots to visualize the distribution of quantitative data
##### Some plots to visualize the distributed data: 
- Distplot/Histplot
- Jointplot
- Pairplot
- Rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Seaborn has it's own numerous datasets, available at seaborn.load_datasets

df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
# this shows that, sex, smoker, day, time and size are qualitative (or categorical) type of data, so these can not be evaluated by distribution plots. We will only use total_bill and tip for that.
```

### DistPlot

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(df['total_bill'], bins = 30, kde = True)
# Bins represents histogram bars, KDE stands for Kernel Density Estimation
# Itâ€™s a way to estimate the probability density function (PDF) of a dataset, i.e. instead of just plotting a histogram with bars, KDE gives you a smooth continuous curve that shows how the data is distributed
```

##### Now making total_bill and tip graphs using matplotlib

```python
plt.subplot(1, 2, 1)
sns.histplot(df['total_bill'], bins = 20, kde = True)
plt.subplot(1, 2, 2)
sns.histplot(df['tip'], bins = 20, kde = True)
```

### Joint plot (for comparing two quantitative distributions)

```python
# Scatter plot

sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'scatter')
```

```python
# Hexagonal plot (like scatter)

sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'hex')
```

```python
# Regression plot

sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
```

```python
# Kernel Density Estimation (KDE) plot

sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'kde')
```

### Pair Plots

```python
# A pair plot pairs all numeric values and give their linear comparasion

sns.pairplot(df)
```

```python
# Pair plots distributed on basis of sex

sns.pairplot(df, hue = 'sex')
```

```python
# Colored pair plots

sns.pairplot(df, hue = 'sex', palette = 'rainbow')
```

### Rug Plots

```python
# Rug-like (like fur on rugs, more fur = more value)
sns.rugplot(df['total_bill'])
```

```python
# Plotting whole dataframe
sns.rugplot(df)
```
